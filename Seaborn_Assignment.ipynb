## Create Visualizations

### Subtask:
Generate visualizations to represent the findings from the analysis steps.

**Reasoning:**
Create a count plot to visualize the distribution of tips by day.

```python
# Visualize tips by day
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=df_tips, order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Distribution of Tips by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount')
plt.show()
```

**Reasoning:**
Create a count plot to visualize the distribution of tips by time.

```python
# Visualize tips by time
plt.figure(figsize=(8, 6))
sns.boxplot(x='time', y='tip', data=df_tips, order=['Lunch', 'Dinner'])
plt.title('Distribution of Tips by Time')
plt.xlabel('Time of Day')
plt.ylabel('Tip Amount')
plt.show()
```

**Reasoning:**
Create a scatter plot to visualize the relationship between tip amount and the size of the party.

```python
# Visualize tips by size of the party
plt.figure(figsize=(8, 6))
sns.scatterplot(x='size', y='tip', data=df_tips)
plt.title('Tip Amount vs. Party Size')
plt.xlabel('Party Size')
plt.ylabel('Tip Amount')
plt.show()
```

**Reasoning:**
Create a scatter plot to visualize the relationship between tip amount and the total bill.

```python
# Visualize tips by total bill
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=df_tips)
plt.title('Tip Amount vs. Total Bill')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.show()
```

**Reasoning:**
Create a box plot to visualize the distribution of tips by smoker status.

```python
# Visualize tips by smoker status
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='tip', data=df_tips)
plt.title('Distribution of Tips by Smoker Status')
plt.xlabel('Smoker')
plt.ylabel('Tip Amount')
plt.show()
```

**Reasoning:**
Create a box plot to visualize the distribution of tips by sex of the payer.

```python
# Visualize tips by sex of the payer
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='tip', data=df_tips)
plt.title('Distribution of Tips by Sex of Payer')
plt.xlabel('Sex')
plt.ylabel('Tip Amount')
plt.show()
```

## Create a dashboard

### Subtask:
Combine the visualizations into a dashboard layout.

```python
# Create a dashboard layout
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each visualization on a separate subplot
sns.boxplot(x='day', y='tip', data=df_tips, order=['Thur', 'Fri', 'Sat', 'Sun'], ax=axes[0])
axes[0].set_title('Distribution of Tips by Day')
axes[0].set_xlabel('Day of the Week')
axes[0].set_ylabel('Tip Amount')

sns.boxplot(x='time', y='tip', data=df_tips, order=['Lunch', 'Dinner'], ax=axes[1])
axes[1].set_title('Distribution of Tips by Time')
axes[1].set_xlabel('Time of Day')
axes[1].set_ylabel('Tip Amount')

sns.scatterplot(x='size', y='tip', data=df_tips, ax=axes[2])
axes[2].set_title('Tip Amount vs. Party Size')
axes[2].set_xlabel('Party Size')
axes[2].set_ylabel('Tip Amount')

sns.scatterplot(x='total_bill', y='tip', data=df_tips, ax=axes[3])
axes[3].set_title('Tip Amount vs. Total Bill')
axes[3].set_xlabel('Total Bill Amount')
axes[3].set_ylabel('Tip Amount')

sns.boxplot(x='smoker', y='tip', data=df_tips, ax=axes[4])
axes[4].set_title('Distribution of Tips by Smoker Status')
axes[4].set_xlabel('Smoker')
axes[4].set_ylabel('Tip Amount')

sns.boxplot(x='gender', y='tip', data=df_tips, ax=axes[5])
axes[5].set_title('Distribution of Tips by Sex of Payer')
axes[5].set_xlabel('Sex')
axes[5].set_ylabel('Tip Amount')


# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Display the dashboard
plt.show()
```
